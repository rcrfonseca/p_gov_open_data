```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
data = pd.read_csv('./data/PoliciaisMortos.csv',sep=';')
```

```python
data.describe()
```

## Perguntas

### 1. Quantos policiais militares mortos em serviço no total? (2003-2018)

```python
pol_mil_mortos_total = data['pol_militares_mortos_serv'].sum()
pol_mil_mortos_total
```

### Gráfico mortes de policiais militares entre 2003-2018

```python
ax = sns.catplot(data=data, 
                 x='vano',
                 y='pol_militares_mortos_serv',
                 kind='bar',
                 aspect=2.5,
                 height=5,
                 palette='plasma',
                 legend_out=True
                 )
ax.set_axis_labels('Ano','Policiais Militares Mortos')
#ax.set(ylim=(0,1))
```

### 2. Quantos policiais civis mortos em serviço no total? (2003-2018)

```python
pol_civ_mortos_total = data['pol_civis_mortos_serv'].sum()
pol_civ_mortos_total
```

### Gráfico mortes de policiais civis entre 2003-2018

```python
ax = sns.catplot(data=data, 
                 x='vano',
                 y='pol_civis_mortos_serv',
                 kind='bar',
                 aspect=2.5,
                 height=5,
                 palette='plasma',
                 legend_out=True
                 )
ax.set_axis_labels('Ano','Policiais Civis Mortos')
```

### 3. Qual índice de morte por mês entre 2003-2018?

### Polícia Militar

```python
pol_mil_mes = data.groupby(['mes','vano'],as_index=False)[['pol_militares_mortos_serv']].agg('sum').pivot('mes','vano')
pol_mil_mes
```

### Polícia Civil

```python
pol_civ_mes = data.groupby(['mes','vano'],as_index=False)[['pol_civis_mortos_serv']].agg('sum').pivot('mes','vano')
pol_civ_mes
```
